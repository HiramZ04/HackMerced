```python
from jetson import get_buffer_text

def infer_text(user_query):
    context = get_buffer_text()  # importarás esto de jetson.py
    response = chat(model='llama3.1:8b', messages=[
        {'role': 'system', 'content': system_prompt},
        {'role': 'user',   'content': f"Nearby: {context}\n\n{user_query}"}
    ])
    return response['message']['content']
```

```python
def infer_image(img, user_query):
    context = get_buffer_text()
    response = chat(model='llava:7b', messages=[
        {
            'role': 'system',
            'content': system_prompt
        },
        {
            'role':'user',
            'content': f"This is the context {context}\n\n{user_query}",
            'images': [img]      
        }
    ])
    return response['message']['content']
    
```

```python
infer_image("../images/classroom.jpg", "Where am I?")
```

```python
infer_text("Can you tell me more about the distance of the objects i have nearby??")
```
